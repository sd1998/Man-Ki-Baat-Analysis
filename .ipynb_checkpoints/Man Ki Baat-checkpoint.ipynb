```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
ps = nltk.PorterStemmer()
summaries = []
for index,row in dataset.iterrows():
    print("Index: ",index)
    text = row['Speech_text']
    words = nltk.word_tokenize(text)
    for i in range(0,len(words)):
        words[i] = ps.stem(words[i]) 
    freq = {}
    for word in words:
        word = word.lower()
        if word not in stopwords:
            freq[word] = freq.get(word,0) + 1
        sentences = nltk.sent_tokenize(text)
        sentf = {}
        for sentence in sentences:
            for word in freq:
                if word in sentence.lower():
                    sentf[sentence[:12]] = sentf.get(sentence[:12],0) + freq[word]
        sumValues = 0
    for sentence in sentf:
        sumValues += sentf[sentence]
    average = int(sumValues/ len(sentf))
    summary = ''
    for sentence in sentences:
        if sentence[:12] in sentf and sentf[sentence[:12]] > (1.5 * average):
            summary +=  " " + sentence
    summaries.append(summary)
```

```python
nltk.download('wordnet')
```

```python
stemmer = nltk.stem.SnowballStemmer('english')
```

```python
def lemmatize(text):
    return stemmer.stem(nltk.stem.WordNetLemmatizer().lemmatize(text))
```

```python
def preprocess(text):
    processed = []
    words = nltk.word_tokenize(text)
    for word in words:
        if word not in stopwords:
            processed.append(lemmatize(word))
    return processed
```

```python
processed_dataset = dataset['Speech_text'].map(preprocess)
```

```python
processed_dataset
```

```python
dictionary = gensim.corpora.Dictionary(processed_dataset)
dictionary.filter_extremes(no_below=15,no_above=0.5,keep_n=100000)
```

```python
#Topic modelling using bag of words 
bow_corpus = [dictionary.doc2bow(data) for data in processed_dataset]
```

```python
lda_bow = gensim.models.LdaMulticore(bow_corpus,num_topics=5,id2word=dictionary,passes=2,workers=2)
```

```python
for index,topic in lda_bow.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(index, topic))
```

```python
#Topic modelling using tfidf
tfidf = gensim.models.TfidfModel(bow_corpus)
tfidf_corpus = tfidf[bow_corpus]
```

```python
lda_tfidf = gensim.models.LdaMulticore(tfidf_corpus,num_topics=5,id2word=dictionary,passes=2,workers=4)
```

```python
for index,topic in lda_tfidf.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(index, topic))
```
